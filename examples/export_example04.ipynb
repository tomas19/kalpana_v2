In this example, we will visualize the maximum flooding levels from a netcdf file <br>
using a mesh from a fort.14 file as the background map. <br>
<br>
This removes dependency on the cartopy package for plotting maximum flooding.

```python
import sys

#import Kalpana functions from github repository
# sys.path.append(r'C:/Users/brand/Github/Kalpana')
from export import *
from visualizations import *
```

**The following cell runs fort14togdf() and stores the output as a geodataframe.** <br>

```python
## path of the input fort 14 file
filein = r'../adds/inputs_examples/fort.14'

## coordinate system of the fort14 input
## Fort 14 almost always uses latitude and longitude
epsgIn = 4326

## coordinate system of the output shapefile
## For this example we use latitude and longitude
epsgOut = 4326

## call fort14togdf() to create a geodataframe object
gdf = fort14togdf(filein, epsgIn, epsgOut, fileintype='fort.14')
```

**The following cells create visualizations topography and flooding levels at specific locations.** <br>
They may take a while to generate.

```python
# North Carolina

# define a figure and axis
fig, ax = plt.subplots(figsize = (8,6), nrows = 1, ncols = 1, subplot_kw={'projection': ccrs.PlateCarree()}, constrained_layout=True)

# visualize the max flooding in water areas
ncFile = r'../adds/inputs_examples/maxele.63.nc'
vis_netcdf(ncFile , 'zeta_max', [0, 3, 0.5], xlims = [-78.5, -75], ylims = [33.5, 37], ax = ax, fig = fig, cbar = True, 
           cbar_label = 'Max water level [m MSL]', ticks = np.arange(-0.25, 3.25, 0.5), background_map = False)

# visualize the mesh to serve as the map for land areas
vis_mesh(gdf, 'zmean', xylims = [-78.5, -75, 33.5, 37], ax = ax, fig = fig, cbar_label = 'Elevation [m MSL]', 
         cmap = 'gray', topo = True, background_map = False)
fig.suptitle('North Carolina Elevation and Flooding', fontsize = 16)
```

```python
# Carolina Beach

# define a figure and axis
fig, ax = plt.subplots(figsize = (8,6), nrows = 1, ncols = 1, subplot_kw={'projection': ccrs.PlateCarree()}, constrained_layout=True)

# visualize the max flooding in water areas
vis_netcdf(ncFile , 'zeta_max', [0, 3, 0.5], xlims = [-78.25, -77.75], ylims = [33.75, 34.25], 
           ax = ax, fig = fig, cbar = True, cbar_label = 'Max water level [m MSL]', ticks = np.arange(-0.25, 3.25, 0.5), background_map = False)

# visualize the mesh to serve as the map for land areas
vis_mesh(gdf, 'zmean', xylims = [-78.25, -77.75, 33.75, 34.25], ax = ax, fig = fig, cbar_label = 'Elevation [m MSL]', 
         cmap = 'gray', topo = True, background_map = False)
fig.suptitle('Carolina Beach Elevation and Flooding', fontsize = 16);
```
